```python
import sqlalchemy 
import pandas as pd 
import os 
```

```python
# Set up the connection 
conn = sqlalchemy.create_engine("postgresql://postgres:1999@localhost/hospital_intelligence")
```

```python
# Cheking the connection status
with conn.connect() as connection:
    print("connected successfully")
```

```python
# loading cleaned dataset 
df = pd.read_csv("/Users/bulbulgupta/Documents/hospital_intelligence/data/processed/cleaned_healthcare.csv")
```

```python
# spliting dataset into two tables 
df_patient = df[["patient_id", "name", "age", "gender", "blood_type"]]
df_admissions = df[["patient_id", "medical_condition", "date_of_admission",
    "doctor", "hospital", "insurance_provider", "billing_amount", "room_number", "admission_type", 
                    "discharge_date", "medication", "test_results", "length_of_stay","readmission_flag"]]
```

```python
# exporting data into postgreSQL database 
df_patient.to_sql('dim_patient', conn, if_exists='replace', index=False)
df_admissions.to_sql('fact_admission', conn, if_exists='replace', index=False)
```
